```python
from pprint import pprint
from vedicastro import horary_chart
from vedicastro.utils import pretty_data_table
from vedicastro.VedicAstro import VedicHoroscopeData
```

```python
## Chart Input - 1
# year = 2024
# month = 2
# day = 5
# hour = 9
# minute = 14 
# secs = 0
# horary_number = 168
```

```python
## Chart Input - 2
year = 2024
month = 2
day = 5
hour = 9
minute = 5
secs = 0
horary_number = 34
```

```python
## General Specs
latitude, longitude, utc = 11.020085773931049, 76.98319647719487, "+5:30" ## Coimbatore
ayan = "Krishnamurti"
house_system = "Placidus" ## Default House for KP system
```

```python
vhd_hora = VedicHoroscopeData(year, month, day, hour, minute, secs, utc, latitude, longitude, ayan, house_system)
vhd_hora_chart = vhd_hora.generate_chart()
horary_asc = horary_chart.get_horary_ascendant_degree(horary_number)
```

```python
desired_asc = horary_asc["ZodiacDegreeLocation"]
print(f'Ascendant Degrees: {desired_asc} degrees')
matched_time, final_chart, planets_data, houses_data = horary_chart.find_exact_ascendant_time(year, month, day, hour, minute, secs,utc,
                                                                    latitude, longitude, horary_asc, house_system, ayan)
asc = houses_data[0]
asc_lon_deg = asc.LonDecDeg

if matched_time:
    print(f'The Ascendant is at {asc_lon_deg} degrees on {matched_time}')
```

```python
print(horary_asc)
pretty_data_table(houses_data)
```

```python
planets_data = [planet for planet in planets_data if planet.Object not in ["Asc", "Chiron", "Syzygy", "Fortuna"]]
pretty_data_table(planets_data)
```

```python
# Splitting the string into date and time
date_part, time_part = matched_time.split(' ')
yy, mon, dd = map(int, date_part.split('-'))
time_elements = time_part.split(':')
hh, mm, ss = int(time_elements[0]), int(time_elements[1]), float(time_elements[2])
print(f"Year: {yy}, Month: {mon}, Day: {dd}, Hour: {hh}, Minute: {mm}, Seconds: {ss}")
```

```python
vhd_hora = VedicHoroscopeData(yy, mon, dd, hh, mm, ss, utc, latitude, longitude, ayan, house_system)
print(f"HouseSystem:{vhd_hora.get_house_system()}; Date: {vhd_hora.year}-{vhd_hora.month}-{vhd_hora.day}; Time: {vhd_hora.hour}:{vhd_hora.minute}:{vhd_hora.second}")
vhd_hora_chart = vhd_hora.generate_chart()
new_houses_data = vhd_hora.get_houses_data_from_chart(vhd_hora_chart)
pretty_data_table(new_houses_data)
```
