```python
from pprint import pprint
from vedicastro import horary_chart
from vedicastro.utils import pretty_data_table, utc_offset_str_to_float
from vedicastro.VedicAstro import VedicHoroscopeData
```

```python
## Chart Input - 1
# year = 2024
# month = 2
# day = 5
# hour = 9
# minute = 14 
# secs = 0
# horary_number = 168
```

```python
## Chart Input - 2
year = 2024
month = 2
day = 5
hour = 9
minute = 5
secs = 0
horary_number = 34
```

```python
## General Specs
latitude, longitude, utc = 11.020085773931049, 76.98319647719487, "+5:30" ## Coimbatore
ayan = "Krishnamurti"
house_system = "Placidus" ## Default House for KP system
tz_offset = utc_offset_str_to_float(utc)
print(tz_offset)
```

```python
horary_asc = horary_chart.get_horary_ascendant_degree(horary_number)
desired_asc = horary_asc["ZodiacDegreeLocation"]
print(f'Desired Asc Degrees: {desired_asc}°')
matched_time  = horary_chart.find_exact_ascendant_time(year, month, day, tz_offset, latitude, longitude, desired_asc, ayan)
secs_final = round(matched_time.second + (matched_time.microsecond) / 1_000_000, 1) + 0.05
print("Rounded Decimal Seconds:",secs_final)
```

```python
vhd_hora_houses = VedicHoroscopeData(year, month, day, matched_time.hour, matched_time.minute, secs_final, utc, latitude, longitude, ayan, "Placidus")
vhd_hora_houses_chart = vhd_hora_houses.generate_chart()
houses_data = vhd_hora_houses.get_houses_data_from_chart(vhd_hora_houses_chart)
```

```python
vhd_hora = VedicHoroscopeData(year, month, day, hour, minute, secs, utc, latitude, longitude, ayan, house_system)
vhd_hora_planets_chart = vhd_hora.generate_chart()
planets_data = vhd_hora.get_planets_data_from_chart(vhd_hora_planets_chart, vhd_hora_houses_chart)
planets_data_filtered = [planet for planet in planets_data if planet.Object not in ["Chiron", "Syzygy", "Fortuna"]]
pretty_data_table(planets_data_filtered)
```

```python
asc = houses_data[0]
if matched_time:
    print(f'The Ascendant is at {asc.LonDecDeg} degrees on {matched_time}')

print(horary_asc)
pretty_data_table(houses_data)    
```
